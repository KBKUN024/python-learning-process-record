````python
# åˆ›å»ºä¸€ä¸ªç®€å•çš„ä¸‰ç»´æ•°ç»„æ¥ç†è§£ç»“æ„
# å½¢çŠ¶ï¼š(3, 4, 5) = (æ·±åº¦, é«˜åº¦, å®½åº¦) = (é¡µæ•°, è¡Œæ•°, åˆ—æ•°)

# æ–¹æ³•1ï¼šç›´æ¥åˆ›å»º
arr_3d = np.arange(60).reshape(3, 4, 5)

print("ğŸ” ä¸‰ç»´æ•°ç»„çš„åŸºæœ¬ä¿¡æ¯ï¼š")
print(f"å½¢çŠ¶ (shape): {arr_3d.shape}")
print(f"ç»´åº¦æ•° (ndim): {arr_3d.ndim}")
print(f"å…ƒç´ æ€»æ•° (size): {arr_3d.size}")
print(f"æ•°æ®ç±»å‹ (dtype): {arr_3d.dtype}")

print("\nğŸ“– å®Œæ•´çš„ä¸‰ç»´æ•°ç»„ï¼š")
print(arr_3d)

print("\nğŸ” è®©æˆ‘ä»¬åˆ†è§£è¿™ä¸ªä¸‰ç»´æ•°ç»„ï¼š")
print("è¿™ä¸ªæ•°ç»„æœ‰3é¡µï¼Œæ¯é¡µ4è¡Œ5åˆ—")

for i in range(3):
    print(f"\nğŸ“„ ç¬¬{i}é¡µ (axis=0, index={i}):")
    print(arr_3d[i])  # è¿™å°±æ˜¯ç¬¬iä¸ªäºŒç»´æ•°ç»„
````

## 2. ä¸‰ä¸ªè½´çš„è¯¦ç»†è§£é‡Š

### ğŸ¯ å…³é”®ç†è§£ï¼šè½´çš„ç¼–å·å’Œå«ä¹‰

åœ¨numpyä¸­ï¼Œä¸‰ç»´æ•°ç»„çš„è½´ç¼–å·æ˜¯ï¼š
- **axis=0ï¼ˆè½´0ï¼‰**ï¼šæ·±åº¦è½´ï¼Œæ§åˆ¶"å“ªä¸€å±‚/é¡µ"
- **axis=1ï¼ˆè½´1ï¼‰**ï¼šè¡Œè½´ï¼Œæ§åˆ¶"å“ªä¸€è¡Œ" 
- **axis=2ï¼ˆè½´2ï¼‰**ï¼šåˆ—è½´ï¼Œæ§åˆ¶"å“ªä¸€åˆ—"

### ğŸŒŸ ä¸ºä»€ä¹ˆç¬¬ä¸‰ä¸ªè½´å«"æ·±åº¦è½´"ï¼Ÿ
è¿™æ˜¯ä¸€ä¸ªçº¦å®šä¿—æˆçš„å«æ³•ï¼š
1. **ä»æ•°å­¦è§’åº¦**ï¼šZè½´é€šå¸¸è¡¨ç¤ºæ·±åº¦/é«˜åº¦
2. **ä»å­˜å‚¨è§’åº¦**ï¼šaxis=0æ˜¯æœ€å¤–å±‚çš„ç´¢å¼•
3. **ä»å¯è§†åŒ–è§’åº¦**ï¼šæƒ³è±¡å¤šä¸ªäºŒç»´å›¾ç‰‡å åœ¨ä¸€èµ·

### ğŸ“ åæ ‡ç³»ç»Ÿï¼š
```
å½¢çŠ¶ (3, 4, 5) è¡¨ç¤ºï¼š
â”œâ”€â”€ axis=0: 3ä¸ª"é¡µé¢"ï¼ˆæ·±åº¦æ–¹å‘ï¼‰
â”œâ”€â”€ axis=1: 4è¡Œï¼ˆå‚ç›´æ–¹å‘ï¼‰  
â””â”€â”€ axis=2: 5åˆ—ï¼ˆæ°´å¹³æ–¹å‘ï¼‰
```


````python
# å¯è§†åŒ–ä¸‰ç»´æ•°ç»„çš„ç»“æ„
def visualize_3d_array_structure():
    """å¯è§†åŒ–ä¸‰ç»´æ•°ç»„çš„è½´å’Œç´¢å¼•"""
    
    fig = plt.figure(figsize=(15, 10))
    
    # åˆ›å»ºä¸€ä¸ªå°çš„ä¸‰ç»´æ•°ç»„ç”¨äºæ¼”ç¤º
    demo_array = np.arange(24).reshape(2, 3, 4)
    
    # ç»˜åˆ¶æ¯ä¸€å±‚
    for i in range(2):
        ax = fig.add_subplot(2, 3, i*3 + 1)
        im = ax.imshow(demo_array[i], cmap='viridis', alpha=0.8)
        ax.set_title(f'ç¬¬{i}å±‚ (axis=0, index={i})')
        ax.set_xlabel('axis=2 (åˆ—)')
        ax.set_ylabel('axis=1 (è¡Œ)')
        
        # æ·»åŠ æ•°å€¼æ ‡æ³¨
        for row in range(3):
            for col in range(4):
                ax.text(col, row, f'{demo_array[i, row, col]}', 
                       ha='center', va='center', color='white', fontsize=10)
    
    # ç»˜åˆ¶è½´çš„è¯´æ˜å›¾
    ax_explain = fig.add_subplot(2, 3, (1, 6))
    ax_explain.text(0.1, 0.8, 'ä¸‰ç»´æ•°ç»„åæ ‡ç³»ç»Ÿ:', fontsize=16, fontweight='bold')
    ax_explain.text(0.1, 0.7, 'â€¢ axis=0 (æ·±åº¦): é€‰æ‹©å“ªä¸€å±‚', fontsize=12, color='red')
    ax_explain.text(0.1, 0.6, 'â€¢ axis=1 (è¡Œ): é€‰æ‹©å“ªä¸€è¡Œ', fontsize=12, color='green') 
    ax_explain.text(0.1, 0.5, 'â€¢ axis=2 (åˆ—): é€‰æ‹©å“ªä¸€åˆ—', fontsize=12, color='blue')
    ax_explain.text(0.1, 0.3, f'æ•°ç»„å½¢çŠ¶: {demo_array.shape}', fontsize=12)
    ax_explain.text(0.1, 0.2, '(æ·±åº¦, è¡Œæ•°, åˆ—æ•°)', fontsize=12, style='italic')
    ax_explain.set_xlim(0, 1)
    ax_explain.set_ylim(0, 1)
    ax_explain.axis('off')
    
    plt.tight_layout()
    plt.show()
    
    return demo_array

demo_array = visualize_3d_array_structure()
print(f"æ¼”ç¤ºæ•°ç»„å½¢çŠ¶: {demo_array.shape}")
print(f"æ¼”ç¤ºæ•°ç»„å†…å®¹:\n{demo_array}")
````

## 4. å®æˆ˜æ¡ˆä¾‹ï¼šRGBå›¾åƒå¤„ç†

### ğŸ–¼ï¸ å›¾åƒå°±æ˜¯ä¸‰ç»´æ•°ç»„ï¼
åœ¨å›¾åƒå¤„ç†ä¸­ï¼Œä¸€å¼ å½©è‰²å›¾ç‰‡å°±æ˜¯ä¸€ä¸ªä¸‰ç»´æ•°ç»„ï¼š
- **axis=0ï¼ˆé«˜åº¦ï¼‰**ï¼šå›¾ç‰‡çš„è¡Œæ•°
- **axis=1ï¼ˆå®½åº¦ï¼‰**ï¼šå›¾ç‰‡çš„åˆ—æ•°  
- **axis=2ï¼ˆé€šé“ï¼‰**ï¼šé¢œè‰²é€šé“ï¼ˆR, G, Bï¼‰

**æ³¨æ„**ï¼šåœ¨å›¾åƒå¤„ç†ä¸­ï¼Œaxisçš„å«ä¹‰ä¸æˆ‘ä»¬å‰é¢çš„ä¾‹å­ç•¥æœ‰ä¸åŒï¼


````python
# åˆ›å»ºä¸€ä¸ªæ¨¡æ‹Ÿçš„RGBå›¾åƒ
def create_sample_image():
    """åˆ›å»ºä¸€ä¸ªå½©è‰²çš„ç¤ºä¾‹å›¾åƒ"""
    height, width = 100, 150
    
    # åˆ›å»ºæ¸å˜å›¾åƒ
    r_channel = np.linspace(0, 255, height*width).reshape(height, width)
    g_channel = np.linspace(255, 0, height*width).reshape(height, width)
    b_channel = np.full((height, width), 128)  # å›ºå®šçš„è“è‰²å€¼
    
    # åˆå¹¶ä¸‰ä¸ªé€šé“
    image = np.stack([r_channel, g_channel, b_channel], axis=2)
    return image.astype(np.uint8)

# åˆ›å»ºç¤ºä¾‹å›¾åƒ
sample_image = create_sample_image()

print("ğŸ–¼ï¸ ç¤ºä¾‹å›¾åƒä¿¡æ¯:")
print(f"å›¾åƒå½¢çŠ¶: {sample_image.shape}")
print(f"æ•°æ®ç±»å‹: {sample_image.dtype}")
print(f"åƒç´ å€¼èŒƒå›´: {sample_image.min()} ~ {sample_image.max()}")

# å¯è§†åŒ–åŸå§‹å›¾åƒå’Œå„ä¸ªé€šé“
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# æ˜¾ç¤ºåŸå§‹å›¾åƒ
axes[0, 0].imshow(sample_image)
axes[0, 0].set_title('åŸå§‹RGBå›¾åƒ')
axes[0, 0].axis('off')

# æ˜¾ç¤ºå„ä¸ªé¢œè‰²é€šé“
channels = ['Red', 'Green', 'Blue']
colors = ['Reds', 'Greens', 'Blues']

for i, (channel_name, colormap) in enumerate(zip(channels, colors)):
    row, col = (0, 1) if i == 0 else (1, i-1)
    axes[row, col].imshow(sample_image[:, :, i], cmap=colormap)
    axes[row, col].set_title(f'{channel_name} é€šé“ (axis=2, index={i})')
    axes[row, col].axis('off')

plt.tight_layout()
plt.show()

print(f"\nğŸ” å›¾åƒæ•°ç»„çš„è¯¦ç»†ä¿¡æ¯:")
print(f"Ré€šé“å½¢çŠ¶: {sample_image[:, :, 0].shape}")
print(f"Gé€šé“å½¢çŠ¶: {sample_image[:, :, 1].shape}")
print(f"Bé€šé“å½¢çŠ¶: {sample_image[:, :, 2].shape}")
````

````python
# å›¾åƒåˆ‡ç‰‡æ“ä½œæ¼”ç¤º
print("ğŸ“¸ å›¾åƒåˆ‡ç‰‡æ“ä½œå®æˆ˜:")
print("="*40)

# 1. æå–å›¾åƒåŒºåŸŸï¼ˆROI - Region of Interestï¼‰
print("\n1ï¸âƒ£ æå–å›¾åƒåŒºåŸŸ:")
roi = sample_image[20:80, 50:120, :]  # æå–ä¸­é—´åŒºåŸŸ
print(f"åŸå›¾å½¢çŠ¶: {sample_image.shape}")
print(f"ROIå½¢çŠ¶: {roi.shape}")

# 2. åˆ†ç¦»é¢œè‰²é€šé“
print("\n2ï¸âƒ£ åˆ†ç¦»é¢œè‰²é€šé“:")
red_channel = sample_image[:, :, 0]    # çº¢è‰²é€šé“
green_channel = sample_image[:, :, 1]  # ç»¿è‰²é€šé“
blue_channel = sample_image[:, :, 2]   # è“è‰²é€šé“

print(f"çº¢è‰²é€šé“å½¢çŠ¶: {red_channel.shape}")

# 3. åˆ›å»ºå•è‰²å›¾åƒ
print("\n3ï¸âƒ£ åˆ›å»ºå•è‰²å›¾åƒ:")
# åªä¿ç•™çº¢è‰²é€šé“ï¼Œå…¶ä»–è®¾ä¸º0
red_only = sample_image.copy()
red_only[:, :, 1] = 0  # ç»¿è‰²é€šé“ç½®é›¶
red_only[:, :, 2] = 0  # è“è‰²é€šé“ç½®é›¶

# 4. å›¾åƒç¿»è½¬
print("\n4ï¸âƒ£ å›¾åƒå˜æ¢:")
flipped_horizontal = sample_image[:, ::-1, :]  # æ°´å¹³ç¿»è½¬
flipped_vertical = sample_image[::-1, :, :]    # å‚ç›´ç¿»è½¬

# 5. å›¾åƒé™é‡‡æ ·
print("\n5ï¸âƒ£ å›¾åƒé™é‡‡æ ·:")
downsampled = sample_image[::2, ::2, :]  # éš”è¡Œéš”åˆ—é‡‡æ ·
print(f"é™é‡‡æ ·åå½¢çŠ¶: {downsampled.shape}")

# å¯è§†åŒ–å„ç§åˆ‡ç‰‡ç»“æœ
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.flatten()

images = [
    (sample_image, "åŸå§‹å›¾åƒ"),
    (roi, "ROIåŒºåŸŸ"),
    (red_channel, "çº¢è‰²é€šé“"),
    (green_channel, "ç»¿è‰²é€šé“"), 
    (blue_channel, "è“è‰²é€šé“"),
    (red_only, "åªä¿ç•™çº¢è‰²"),
    (flipped_horizontal, "æ°´å¹³ç¿»è½¬"),
    (flipped_vertical, "å‚ç›´ç¿»è½¬"),
    (downsampled, "é™é‡‡æ ·")
]

for i, (img, title) in enumerate(images):
    if len(img.shape) == 2:  # ç°åº¦å›¾åƒ
        axes[i].imshow(img, cmap='gray')
    else:  # å½©è‰²å›¾åƒ
        axes[i].imshow(img)
    axes[i].set_title(title)
    axes[i].axis('off')

plt.tight_layout()
plt.show()
````

````python
# æ¼”ç¤ºä¸åŒé¢†åŸŸçš„ä¸‰ç»´æ•°ç»„å®ä¾‹
print("ğŸŒ ä¸åŒé¢†åŸŸçš„ä¸‰ç»´æ•°ç»„å®ä¾‹:")
print("="*50)

# 1. ç§‘å­¦æ•°æ®ï¼š3Dæ¸©åº¦åœº
print("\n1ï¸âƒ£ ç§‘å­¦æ•°æ® - 3Dæ¸©åº¦åœº:")
# æ¨¡æ‹Ÿä¸€ä¸ª3Dç©ºé—´ä¸­çš„æ¸©åº¦åˆ†å¸ƒ (Z, Y, X)
z_size, y_size, x_size = 5, 8, 10
temperature_field = np.random.normal(20, 5, (z_size, y_size, x_size))  # å¹³å‡æ¸©åº¦20åº¦

print(f"æ¸©åº¦åœºå½¢çŠ¶: {temperature_field.shape} (Z, Y, X)")
print(f"æ¸©åº¦èŒƒå›´: {temperature_field.min():.1f}Â°C ~ {temperature_field.max():.1f}Â°C")

# æå–æŸä¸ªé«˜åº¦å±‚çš„æ¸©åº¦åˆ†å¸ƒ
z_layer = temperature_field[2, :, :]  # ç¬¬2å±‚çš„æ¸©åº¦åˆ†å¸ƒ
print(f"ç¬¬2å±‚(Z=2)æ¸©åº¦åˆ†å¸ƒå½¢çŠ¶: {z_layer.shape}")

# 2. è§†é¢‘æ•°æ®ï¼šæ—¶é—´åºåˆ—å›¾åƒ
print("\n2ï¸âƒ£ è§†é¢‘æ•°æ® - æ—¶é—´åºåˆ—:")
# æ¨¡æ‹Ÿè§†é¢‘å¸§ (æ—¶é—´, é«˜åº¦, å®½åº¦)
frames, height, width = 10, 64, 64
video_data = np.random.randint(0, 255, (frames, height, width), dtype=np.uint8)

print(f"è§†é¢‘æ•°æ®å½¢çŠ¶: {video_data.shape} (æ—¶é—´, é«˜åº¦, å®½åº¦)")
print(f"æ€»å¸§æ•°: {frames}, æ¯å¸§å°ºå¯¸: {height}Ã—{width}")

# æå–ç¬¬5å¸§
frame_5 = video_data[5, :, :]
print(f"ç¬¬5å¸§å½¢çŠ¶: {frame_5.shape}")

# 3. æ·±åº¦å­¦ä¹ ï¼šæ‰¹é‡å›¾åƒæ•°æ®
print("\n3ï¸âƒ£ æ·±åº¦å­¦ä¹  - æ‰¹é‡æ•°æ®:")
# æ¨¡æ‹Ÿä¸€ä¸ªå°æ‰¹é‡çš„ç°åº¦å›¾åƒ (æ‰¹æ¬¡, é«˜åº¦, å®½åº¦)
batch_size, img_h, img_w = 4, 32, 32
batch_images = np.random.rand(batch_size, img_h, img_w)

print(f"æ‰¹é‡å›¾åƒå½¢çŠ¶: {batch_images.shape} (æ‰¹æ¬¡, é«˜åº¦, å®½åº¦)")
print(f"æ‰¹æ¬¡å¤§å°: {batch_size}, æ¯å¼ å›¾ç‰‡: {img_h}Ã—{img_w}")

# æå–æ‰¹æ¬¡ä¸­çš„ç¬¬ä¸€å¼ å›¾ç‰‡
first_image = batch_images[0, :, :]
print(f"ç¬¬ä¸€å¼ å›¾ç‰‡å½¢çŠ¶: {first_image.shape}")

# å¯è§†åŒ–è¿™äº›ä¸åŒç±»å‹çš„æ•°æ®
fig, axes = plt.subplots(2, 3, figsize=(15, 10))

# æ¸©åº¦åœºå¯è§†åŒ–
im1 = axes[0, 0].imshow(z_layer, cmap='coolwarm', aspect='auto')
axes[0, 0].set_title('æ¸©åº¦åœº Z=2å±‚')
axes[0, 0].set_xlabel('Xåæ ‡')
axes[0, 0].set_ylabel('Yåæ ‡')
plt.colorbar(im1, ax=axes[0, 0], shrink=0.8)

# è§†é¢‘å¸§å¯è§†åŒ–
axes[0, 1].imshow(frame_5, cmap='gray')
axes[0, 1].set_title('è§†é¢‘ç¬¬5å¸§')
axes[0, 1].axis('off')

# æ‰¹é‡å›¾åƒå¯è§†åŒ–
axes[0, 2].imshow(first_image, cmap='viridis')
axes[0, 2].set_title('æ‰¹æ¬¡ä¸­ç¬¬ä¸€å¼ å›¾ç‰‡')
axes[0, 2].axis('off')

# ä¸‹æ’æ˜¾ç¤ºåˆ‡ç‰‡æ“ä½œç¤ºä¾‹
# æ¸©åº¦åœºçš„å‚ç›´åˆ‡ç‰‡
vertical_slice = temperature_field[:, 4, :]  # Y=4çš„å‚ç›´åˆ‡ç‰‡
im2 = axes[1, 0].imshow(vertical_slice, cmap='coolwarm', aspect='auto')
axes[1, 0].set_title('æ¸©åº¦åœºå‚ç›´åˆ‡ç‰‡ (Y=4)')
axes[1, 0].set_xlabel('Xåæ ‡')
axes[1, 0].set_ylabel('Zåæ ‡')
plt.colorbar(im2, ax=axes[1, 0], shrink=0.8)

# è§†é¢‘æ—¶é—´åˆ‡ç‰‡
time_slice = video_data[:, 32, 32]  # ä¸­å¿ƒåƒç´ çš„æ—¶é—´å˜åŒ–
axes[1, 1].plot(time_slice, marker='o')
axes[1, 1].set_title('ä¸­å¿ƒåƒç´ æ—¶é—´å˜åŒ–')
axes[1, 1].set_xlabel('æ—¶é—´å¸§')
axes[1, 1].set_ylabel('åƒç´ å€¼')
axes[1, 1].grid(True)

# æ‰¹é‡æ•°æ®çš„å¹³å‡å›¾åƒ
mean_image = batch_images.mean(axis=0)  # æ²¿æ‰¹æ¬¡ç»´åº¦æ±‚å¹³å‡
axes[1, 2].imshow(mean_image, cmap='viridis')
axes[1, 2].set_title('æ‰¹æ¬¡å¹³å‡å›¾åƒ')
axes[1, 2].axis('off')

plt.tight_layout()
plt.show()
````

````python
# å®æˆ˜ç»ƒä¹ 
print("ğŸ¯ ä¸‰ç»´æ•°ç»„åˆ‡ç‰‡å®æˆ˜ç»ƒä¹ ")
print("="*50)

# åˆ›å»ºç»ƒä¹ ç”¨çš„æ•°æ®
np.random.seed(42)  # å›ºå®šéšæœºç§å­ï¼Œç¡®ä¿ç»“æœå¯é‡ç°

# ç»ƒä¹ 1ï¼šRGBå›¾åƒæ“ä½œ
print("\nğŸ“¸ ç»ƒä¹ 1: RGBå›¾åƒå¤„ç†")
print("-" * 30)

# åˆ›å»ºä¸€ä¸ªæ¨¡æ‹Ÿçš„RGBå›¾åƒ (128x128x3)
image_rgb = np.random.randint(0, 255, (128, 128, 3), dtype=np.uint8)
print(f"åŸå§‹å›¾åƒå½¢çŠ¶: {image_rgb.shape}")

# ä»»åŠ¡1.1: æå–ç»¿è‰²é€šé“
green_channel = image_rgb[:, :, 1]
print(f"ç»¿è‰²é€šé“å½¢çŠ¶: {green_channel.shape}")

# ä»»åŠ¡1.2: æå–å›¾åƒçš„å·¦ä¸Šè§’64x64åŒºåŸŸ
top_left = image_rgb[:64, :64, :]
print(f"å·¦ä¸Šè§’åŒºåŸŸå½¢çŠ¶: {top_left.shape}")

# ä»»åŠ¡1.3: åˆ›å»ºåªæœ‰çº¢è‰²çš„å›¾åƒ
red_only_image = image_rgb.copy()
red_only_image[:, :, 1] = 0  # ç»¿è‰²é€šé“ç½®é›¶
red_only_image[:, :, 2] = 0  # è“è‰²é€šé“ç½®é›¶

# ä»»åŠ¡1.4: å›¾åƒä¸‹é‡‡æ · (æ¯éš”ä¸€ä¸ªåƒç´ å–æ ·)
downsampled = image_rgb[::2, ::2, :]
print(f"ä¸‹é‡‡æ ·åå½¢çŠ¶: {downsampled.shape}")

# ç»ƒä¹ 2ï¼šè§†é¢‘æ•°æ®å¤„ç†
print("\nğŸ¬ ç»ƒä¹ 2: è§†é¢‘æ•°æ®å¤„ç†")
print("-" * 30)

# åˆ›å»ºæ¨¡æ‹Ÿè§†é¢‘æ•°æ® (30å¸§, 64x64åƒç´ )
video_frames = np.random.randint(0, 255, (30, 64, 64), dtype=np.uint8)
print(f"è§†é¢‘æ•°æ®å½¢çŠ¶: {video_frames.shape}")

# ä»»åŠ¡2.1: æå–ç¬¬10-20å¸§
clip = video_frames[10:21, :, :]
print(f"è§†é¢‘ç‰‡æ®µå½¢çŠ¶: {clip.shape}")

# ä»»åŠ¡2.2: æå–æ‰€æœ‰å¸§çš„ä¸­å¿ƒ16x16åŒºåŸŸ
center_start = 24  # 64//2 - 8
center_end = 40    # 64//2 + 8
center_regions = video_frames[:, center_start:center_end, center_start:center_end]
print(f"ä¸­å¿ƒåŒºåŸŸå½¢çŠ¶: {center_regions.shape}")

# ä»»åŠ¡2.3: æå–æ¯5å¸§çš„æ•°æ®
every_5th_frame = video_frames[::5, :, :]
print(f"æ¯5å¸§å–æ ·å½¢çŠ¶: {every_5th_frame.shape}")

# ç»ƒä¹ 3ï¼šç§‘å­¦æ•°æ®åˆ†æ
print("\nğŸ”¬ ç»ƒä¹ 3: 3Dç§‘å­¦æ•°æ®")
print("-" * 30)

# åˆ›å»º3Dæ¸©åº¦åœºæ•°æ® (10x20x30)
temperature_3d = np.random.normal(25, 10, (10, 20, 30))  # æ¸©åº¦æ•°æ®
print(f"3Dæ¸©åº¦åœºå½¢çŠ¶: {temperature_3d.shape} (Z, Y, X)")

# ä»»åŠ¡3.1: æå–è¡¨é¢å±‚(z=0)çš„æ¸©åº¦åˆ†å¸ƒ
surface_temp = temperature_3d[0, :, :]
print(f"è¡¨é¢æ¸©åº¦åˆ†å¸ƒå½¢çŠ¶: {surface_temp.shape}")

# ä»»åŠ¡3.2: æå–å‚ç›´å‰–é¢(y=10)
vertical_section = temperature_3d[:, 10, :]
print(f"å‚ç›´å‰–é¢å½¢çŠ¶: {vertical_section.shape}")

# ä»»åŠ¡3.3: æå–ç«‹æ–¹ä½“å­åŒºåŸŸ (z:2-8, y:5-15, x:10-25)
cube_region = temperature_3d[2:8, 5:15, 10:25]
print(f"ç«‹æ–¹ä½“åŒºåŸŸå½¢çŠ¶: {cube_region.shape}")

# ä»»åŠ¡3.4: è®¡ç®—æ¯å±‚çš„å¹³å‡æ¸©åº¦
layer_means = temperature_3d.mean(axis=(1, 2))  # æ²¿Yå’ŒXè½´æ±‚å¹³å‡
print(f"æ¯å±‚å¹³å‡æ¸©åº¦: {layer_means.shape}")
print(f"å‰5å±‚å¹³å‡æ¸©åº¦: {layer_means[:5]}")

# å¯è§†åŒ–ç»ƒä¹ ç»“æœ
fig, axes = plt.subplots(3, 4, figsize=(16, 12))

# ç¬¬ä¸€è¡Œï¼šå›¾åƒå¤„ç†ç»“æœ
axes[0, 0].imshow(image_rgb[:64, :64, :])
axes[0, 0].set_title('åŸå›¾(å·¦ä¸Šè§’)')
axes[0, 0].axis('off')

axes[0, 1].imshow(green_channel, cmap='gray')
axes[0, 1].set_title('ç»¿è‰²é€šé“')
axes[0, 1].axis('off')

axes[0, 2].imshow(red_only_image[:64, :64, :])
axes[0, 2].set_title('åªä¿ç•™çº¢è‰²')
axes[0, 2].axis('off')

axes[0, 3].imshow(downsampled[:32, :32, :])
axes[0, 3].set_title('ä¸‹é‡‡æ ·')
axes[0, 3].axis('off')

# ç¬¬äºŒè¡Œï¼šè§†é¢‘å¤„ç†ç»“æœ
axes[1, 0].imshow(video_frames[0], cmap='gray')
axes[1, 0].set_title('ç¬¬0å¸§')
axes[1, 0].axis('off')

axes[1, 1].imshow(video_frames[15], cmap='gray')
axes[1, 1].set_title('ç¬¬15å¸§')
axes[1, 1].axis('off')

axes[1, 2].imshow(center_regions[0], cmap='gray')
axes[1, 2].set_title('ä¸­å¿ƒåŒºåŸŸ(ç¬¬0å¸§)')
axes[1, 2].axis('off')

# æ˜¾ç¤ºæ—¶é—´åºåˆ—
pixel_timeline = video_frames[:, 32, 32]  # ä¸­å¿ƒåƒç´ çš„æ—¶é—´å˜åŒ–
axes[1, 3].plot(pixel_timeline)
axes[1, 3].set_title('ä¸­å¿ƒåƒç´ æ—¶é—´å˜åŒ–')
axes[1, 3].set_xlabel('å¸§æ•°')
axes[1, 3].set_ylabel('åƒç´ å€¼')
axes[1, 3].grid(True)

# ç¬¬ä¸‰è¡Œï¼šç§‘å­¦æ•°æ®ç»“æœ
im1 = axes[2, 0].imshow(surface_temp, cmap='coolwarm')
axes[2, 0].set_title('è¡¨é¢æ¸©åº¦åˆ†å¸ƒ')
plt.colorbar(im1, ax=axes[2, 0], shrink=0.8)

im2 = axes[2, 1].imshow(vertical_section, cmap='coolwarm')
axes[2, 1].set_title('å‚ç›´å‰–é¢(Y=10)')
axes[2, 1].set_xlabel('X')
axes[2, 1].set_ylabel('Z')
plt.colorbar(im2, ax=axes[2, 1], shrink=0.8)

im3 = axes[2, 2].imshow(cube_region[3, :, :], cmap='coolwarm')  # æ˜¾ç¤ºç«‹æ–¹ä½“ä¸­çš„ä¸€å±‚
axes[2, 2].set_title('ç«‹æ–¹ä½“åŒºåŸŸ(Z=5å±‚)')
plt.colorbar(im3, ax=axes[2, 2], shrink=0.8)

axes[2, 3].plot(layer_means, marker='o')
axes[2, 3].set_title('æ¯å±‚å¹³å‡æ¸©åº¦')
axes[2, 3].set_xlabel('å±‚æ•°(Z)')
axes[2, 3].set_ylabel('æ¸©åº¦(Â°C)')
axes[2, 3].grid(True)

plt.tight_layout()
plt.show()

print("\nâœ… ç»ƒä¹ å®Œæˆï¼ä½ å·²ç»æŒæ¡äº†ä¸‰ç»´æ•°ç»„çš„åŸºæœ¬åˆ‡ç‰‡æ“ä½œã€‚")
````
