`è¯·è®°ä½ï¼Œåªæœ‰åˆ†ç»„æ•è·çš„æ—¶å€™ï¼ŒåŒ¹é…çš„ç»“æœæ‰èƒ½ä½¿ç”¨group,åè¿‡æ¥è¯´ï¼Œä½ éƒ½æ²¡åˆ†ç»„ï¼Œgroupä¸ªå±å•ŠğŸ˜‚`
# 1. åˆ†ç»„æ•è·çš„åŸºæœ¬è¯­æ³•â€‹â€‹
â€‹â€‹(1) ç®€å•åˆ†ç»„â€‹â€‹ï¼šç”¨ (pattern) åŒ…è£¹éœ€è¦æ•è·çš„éƒ¨åˆ†ï¼š

```python
import re

text = "Date: 2023-10-05"
pattern = r"Date: (\d{4})-(\d{2})-(\d{2})"  # æ•è·å¹´ã€æœˆã€æ—¥

match = re.search(pattern, text)
if match:
    print(match.groups())  # è¾“å‡º: ('2023', '10', '05')
    print(match.group(0)) # ç´¢å¼•ä¸º0æ—¶æ˜¯æ•´ä¸ªåŒ¹é…çš„æ–‡æœ¬
    print(match.group(1))
    print(match.group(2))
    print(match.group(3))
```

â€‹â€‹- match.groups()â€‹â€‹ï¼šè¿”å›æ‰€æœ‰æ•è·ç»„çš„å…ƒç»„ï¼ˆæŒ‰é¡ºåºï¼‰ã€‚

â€‹â€‹- match.group(n)â€‹â€‹ï¼šè·å–ç¬¬ n ä¸ªåˆ†ç»„ï¼ˆgroup(0) æ˜¯æ•´ä¸ªåŒ¹é…æ–‡æœ¬ï¼‰ã€‚

ä¸Šè¿°ä¾‹å­ä¸­ï¼Œå› ä¸ºæ˜¯åˆ†ç»„æ•è·ï¼Œåˆ†ç»„åˆ†ç»„ï¼Œè‚¯å®šæ˜¯è¢«åˆ†æˆäº†ä¸€ç»„ç»„çš„ï¼Œæ‰€ä»¥å¯ä»¥é€šè¿‡group(n)æ¥è®¿é—®ï¼ŒğŸ”è¯¦è§ä¾‹å­

# 2. åˆ†ç»„æ•è·çš„å¸¸è§ç”¨é€”â€‹â€‹
â€‹â€‹(1) æå–å…³é”®ä¿¡æ¯â€‹

```python
text = "Name: Alice, Age: 25"
match = re.search(r"Name: (\w+), Age: (\d+)", text)
print(match.groups())  # è¾“å‡º: ('Alice', '25')
print(match.group(1)) 
print(match.group(2))  
```

(2) ç»“åˆ findall æ‰¹é‡æå–â€‹

```python
text = "user_id=123, user_id=456, user_id=789"
ids = re.findall(r"user_id=(\d+)", text)  # æ³¨æ„æ‹¬å·ä½ç½®
print(ids)  # è¾“å‡º: ['123', '456', '789']
```

(3) æ›¿æ¢æ—¶å¼•ç”¨åˆ†ç»„ï¼ˆåå‘å¼•ç”¨ï¼‰â€‹

```python
text = "2023-10-05"
new_text = re.sub(r"(\d{4})-(\d{2})-(\d{2})", r"\2/\3/\1", text)  # æœˆ/æ—¥/å¹´
print(new_text)  # è¾“å‡º: 10/05/2023
```

# 3. éæ•è·åˆ†ç»„ï¼ˆé«˜çº§ç”¨æ³•ï¼‰â€‹â€‹
å¦‚æœåªéœ€è¦åˆ†ç»„ä½†ä¸æ•è·ï¼ˆä¸å ç”¨ groups() çš„è¿”å›ï¼‰ï¼Œå¯ä»¥ç”¨ (?:pattern)ï¼š

```python
text = "apple orange banana strawberry"
match = re.findall(r"(?:\w+)(\w+)", text)  # ç¬¬ä¸€ä¸ª `\w+` ä¸æ•è·
print(match)  # è¾“å‡º: ['e', 'e', 'a', 'y']
match2 = re.search(r"(?:\w+) (\w+) (\w+)", text)  # ç¬¬ä¸€ä¸ª `\w+` ä¸æ•è·
print(match2.groups())  # è¾“å‡º: ('orange', 'banana')
```

å› ä¸º (?:\w+) ä¼šè´ªå©ªåŒ¹é…ï¼Œå…ˆåŒ¹é…å°½å¯èƒ½å¤šçš„å­—ç¬¦ï¼Œåªç•™ä¸‹æœ€åä¸€ä¸ªå­—ç¬¦ç»™åé¢çš„ (\w+) æ•è·ã€‚æ¯”å¦‚å¯¹äº "apple"ï¼š

# 4. å‘½ååˆ†ç»„ï¼ˆæ›´æ˜“è¯»ï¼‰â€‹â€‹
é€šè¿‡ (?P<name>pattern) ç»™åˆ†ç»„å‘½åï¼Œä¾¿äºåç»­å¼•ç”¨ï¼š

```python
text = "Date: 2023-10-05"
pattern = r"Date: (?P<year>\d{4})-(?P<month>\d{2})-(?P<day>\d{2})"

match = re.search(pattern, text)
print(match.group("year"))   # è¾“å‡º: '2023'
print(match.groupdict())     # è¾“å‡º: {'year': '2023', 'month': '10', 'day': '05'}
```

# 5. åˆ†ç»„ä¸é‡è¯çš„ç»“åˆâ€‹â€‹
åˆ†ç»„å¯ä»¥é…åˆé‡è¯ï¼ˆ*ã€+ã€?ã€{n,m}ï¼‰ä½¿ç”¨ï¼š

```python
text = "abbbcc"
match = re.search(r"(a)(b+)(c+)", text)
print(match.groups())  # è¾“å‡º: ('a', 'bbb', 'cc')
```

â€‹â€‹(2) è´ªå©ªåŒ¹é…é—®é¢˜â€‹â€‹ï¼šé»˜è®¤æƒ…å†µä¸‹ï¼Œåˆ†ç»„æ˜¯è´ªå©ªçš„ï¼ˆåŒ¹é…å°½å¯èƒ½å¤šçš„å­—ç¬¦ï¼‰ï¼š
