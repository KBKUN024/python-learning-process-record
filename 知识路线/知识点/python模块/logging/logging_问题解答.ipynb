```python
"""
é—®é¢˜1: basicConfig(level=logging.INFO) è¿˜èƒ½æ€ä¹ˆé…ç½®ï¼Ÿ
é—®é¢˜2: ä¸ºä»€ä¹ˆç”¨ getLogger(__name__)ï¼Ÿ
"""

import logging
import sys
from pathlib import Path

print("ğŸ¯ PyBackendPro ç²¾å‡†è§£ç­”ä½ çš„é—®é¢˜")
print("="*60)

# =====================================
# é—®é¢˜1ï¼šlevel å‚æ•°çš„å¤šç§é…ç½®æ–¹å¼
# =====================================
print("\nğŸ“‹ é—®é¢˜1: level=logging.INFO è¿˜èƒ½æ€ä¹ˆé…ç½®ï¼Ÿ")
print("-" * 40)

print("âœ… ç­”æ¡ˆï¼šlevel æœ‰5ä¸ªæ ‡å‡†çº§åˆ« + å¤šç§é…ç½®æ–¹å¼")

# æ–¹å¼1ï¼šä½¿ç”¨å¸¸é‡ï¼ˆæ¨èï¼‰
print("\n1ï¸âƒ£ ä½¿ç”¨ logging å¸¸é‡ (æœ€ä½³å®è·µ):")
config_examples = [
    ("logging.DEBUG", "å¼€å‘è°ƒè¯•", "æ˜¾ç¤ºæ‰€æœ‰ä¿¡æ¯"),
    ("logging.INFO", "ç”Ÿäº§ç›‘æ§", "æ˜¾ç¤ºé‡è¦ä¿¡æ¯"),
    ("logging.WARNING", "è­¦å‘Šæ¨¡å¼", "åªæ˜¾ç¤ºé—®é¢˜"),
    ("logging.ERROR", "é”™è¯¯è¿½è¸ª", "åªæ˜¾ç¤ºé”™è¯¯"),
    ("logging.CRITICAL", "å´©æºƒåˆ†æ", "åªæ˜¾ç¤ºè‡´å‘½é”™è¯¯")
]

for level_code, usage, description in config_examples:
    print(f"   {level_code:20} - {usage:8} - {description}")

# æ–¹å¼2ï¼šä½¿ç”¨æ•°å­—
print("\n2ï¸âƒ£ ä½¿ç”¨æ•°å­—å€¼ (ä¸æ¨èï¼Œä½†åˆæ³•):")
logging.basicConfig(level=20, format='[%(levelname)s] %(message)s', force=True)
logger = logging.getLogger('number_demo')
logger.info("ä½¿ç”¨æ•°å­—20é…ç½®çš„INFOçº§åˆ«")

# æ–¹å¼3ï¼šä½¿ç”¨å­—ç¬¦ä¸²
print("\n3ï¸âƒ£ ä½¿ç”¨å­—ç¬¦ä¸² (åŠ¨æ€é…ç½®å¸¸ç”¨):")
logging.basicConfig(level='WARNING', format='[%(levelname)s] %(message)s', force=True)
logger = logging.getLogger('string_demo')
logger.warning("ä½¿ç”¨å­—ç¬¦ä¸²'WARNING'é…ç½®")

# æ¼”ç¤ºè¾“å‡ºæ ¼å¼çš„å¤šæ ·æ€§
print("\n4ï¸âƒ£ é…ç½®è¾“å‡ºæ ¼å¼çš„èŠ±æ ·:")

# ç®€å•æ ¼å¼
print("\n   ç®€å•æ ¼å¼:")
logging.basicConfig(
    level=logging.INFO,
    format='%(levelname)s: %(message)s',
    force=True
)
simple_logger = logging.getLogger('simple')
simple_logger.info("ç®€å•æ ¼å¼çš„æ¶ˆæ¯")

# è¯¦ç»†æ ¼å¼
print("\n   è¯¦ç»†æ ¼å¼ (ç”Ÿäº§ç¯å¢ƒæ¨è):")
logging.basicConfig(
    level=logging.INFO,
    format='[%(asctime)s] [%(process)d] [%(name)s:%(lineno)d] [%(levelname)s] %(message)s',
    datefmt='%Y-%m-%d %H:%M:%S',
    force=True
)
detailed_logger = logging.getLogger('detailed')
detailed_logger.info("è¯¦ç»†æ ¼å¼åŒ…å«æ—¶é—´ã€è¿›ç¨‹IDã€æ¨¡å—åã€è¡Œå·")

# JSONæ ¼å¼ (ç°ä»£å¾®æœåŠ¡)
print("\n   JSONæ ¼å¼ (å¾®æœåŠ¡/äº‘åŸç”Ÿ):")
import json
from datetime import datetime

class JSONFormatter(logging.Formatter):
    def format(self, record):
        log_entry = {
            "timestamp": datetime.fromtimestamp(record.created).isoformat(),
            "level": record.levelname,
            "logger": record.name,
            "message": record.getMessage(),
            "module": record.module,
            "line": record.lineno
        }
        return json.dumps(log_entry, ensure_ascii=False)

json_handler = logging.StreamHandler()
json_handler.setFormatter(JSONFormatter())
json_logger = logging.getLogger('json_demo')
json_logger.handlers.clear()
json_logger.addHandler(json_handler)
json_logger.setLevel(logging.INFO)
json_logger.info("JSONæ ¼å¼ä¾¿äºæ—¥å¿—åˆ†æå·¥å…·å¤„ç†")

# å½©è‰²è¾“å‡º (å¼€å‘ç¯å¢ƒå‹å¥½)
print("\n   å½©è‰²è¾“å‡º (å¼€å‘ç¯å¢ƒ):")
class SimpleColorFormatter(logging.Formatter):
    COLORS = {
        'DEBUG': '\033[96m',    # äº®é’è‰²
        'INFO': '\033[92m',     # äº®ç»¿è‰²
        'WARNING': '\033[93m',  # äº®é»„è‰²
        'ERROR': '\033[91m',    # äº®çº¢è‰²
        'CRITICAL': '\033[95m', # äº®ç´«è‰²
        'RESET': '\033[0m'
    }
    
    def format(self, record):
        color = self.COLORS.get(record.levelname, '')
        reset = self.COLORS['RESET']
        record.msg = f"{color}{record.msg}{reset}"
        return super().format(record)

color_handler = logging.StreamHandler()
color_handler.setFormatter(SimpleColorFormatter('%(levelname)s: %(message)s'))
color_logger = logging.getLogger('color_demo')
color_logger.handlers.clear()
color_logger.addHandler(color_handler)
color_logger.setLevel(logging.DEBUG)

color_logger.debug("è°ƒè¯•ä¿¡æ¯ (äº®é’è‰²)")
color_logger.info("æ­£å¸¸ä¿¡æ¯ (äº®ç»¿è‰²)")
color_logger.warning("è­¦å‘Šä¿¡æ¯ (äº®é»„è‰²)")
color_logger.error("é”™è¯¯ä¿¡æ¯ (äº®çº¢è‰²)")
color_logger.critical("ä¸¥é‡é”™è¯¯ (äº®ç´«è‰²)")

# =====================================
# é—®é¢˜2ï¼šä¸ºä»€ä¹ˆä½¿ç”¨ getLogger(__name__)
# =====================================
print(f"\n{'='*60}")
print("ğŸ“‹ é—®é¢˜2: ä¸ºä»€ä¹ˆç”¨ getLogger(__name__)ï¼Ÿ")
print("-" * 40)

print("âœ… ç­”æ¡ˆï¼šæ¨¡å—åŒ–ç®¡ç† + å±‚æ¬¡ç»“æ„ + ä¾¿äºè¿½è¸ª")

# æ¼”ç¤º __name__ çš„å®é™…å€¼
print(f"\nğŸ” å½“å‰ __name__ çš„å€¼: '{__name__}'")

# æ¨¡æ‹Ÿä¸åŒæ¨¡å—çš„ __name__
module_examples = [
    ("__main__", "ç›´æ¥è¿è¡Œçš„è„šæœ¬"),
    ("myapp.models.user", "ç”¨æˆ·æ¨¡å‹æ¨¡å—"),
    ("myapp.api.auth", "è®¤è¯APIæ¨¡å—"),
    ("myapp.utils.database", "æ•°æ®åº“å·¥å…·æ¨¡å—"),
    ("third_party.requests", "ç¬¬ä¸‰æ–¹åº“æ¨¡å—")
]

print("\nğŸ“š ä¸åŒæ¨¡å—çš„ __name__ ç¤ºä¾‹:")
for name, description in module_examples:
    print(f"   {name:20} -> {description}")

# æ¼”ç¤ºå±‚æ¬¡åŒ–ç®¡ç†çš„å¨åŠ›
print("\nğŸŒ³ å±‚æ¬¡åŒ–ç®¡ç†æ¼”ç¤º:")

# é‡ç½®loggingé…ç½®
for handler in logging.root.handlers[:]:
    logging.root.removeHandler(handler)

logging.basicConfig(
    level=logging.DEBUG,
    format='[%(name)25s] [%(levelname)8s] %(message)s',
    force=True
)

# åˆ›å»ºå±‚æ¬¡åŒ–çš„logger
app_logger = logging.getLogger('myapp')
db_logger = logging.getLogger('myapp.database')
api_logger = logging.getLogger('myapp.api')
auth_logger = logging.getLogger('myapp.api.auth')
user_logger = logging.getLogger('myapp.models.user')

# è®¾ç½®ä¸åŒçº§åˆ«
app_logger.setLevel(logging.INFO)      # åº”ç”¨çº§åˆ«ï¼šINFOä»¥ä¸Š
db_logger.setLevel(logging.WARNING)    # æ•°æ®åº“ï¼šåªçœ‹è­¦å‘Š
api_logger.setLevel(logging.DEBUG)     # APIï¼šè¯¦ç»†è°ƒè¯•
auth_logger.setLevel(logging.INFO)     # è®¤è¯ï¼šé‡è¦ä¿¡æ¯

print("\nğŸ¯ åŒä¸€ä¸ªæ¶ˆæ¯ï¼Œä¸åŒæ¨¡å—çš„å¤„ç†:")
app_logger.debug("åº”ç”¨è°ƒè¯•ä¿¡æ¯ (ä¸æ˜¾ç¤ºï¼Œçº§åˆ«ä¸å¤Ÿ)")
app_logger.info("åº”ç”¨å¯åŠ¨æˆåŠŸ âœ…")

db_logger.debug("æ•°æ®åº“è¿æ¥æ± åˆå§‹åŒ– (ä¸æ˜¾ç¤º)")
db_logger.info("æ‰§è¡ŒSQLæŸ¥è¯¢ (ä¸æ˜¾ç¤º)")
db_logger.warning("è¿æ¥æ± ä½¿ç”¨ç‡80% âš ï¸")

api_logger.debug("APIè¯·æ±‚è¯¦æƒ… ğŸ”")
api_logger.info("å¤„ç†ç”¨æˆ·è¯·æ±‚ ğŸ“")

auth_logger.debug("éªŒè¯JWT token (ä¸æ˜¾ç¤º)")
auth_logger.info("ç”¨æˆ·ç™»å½•æˆåŠŸ ğŸ”")

user_logger.info("åˆ›å»ºæ–°ç”¨æˆ·è®°å½• ğŸ‘¤")

# æ¼”ç¤ºä¾¿äºè°ƒè¯•çš„å¨åŠ›
print("\nğŸ› ä¾¿äºè°ƒè¯•çš„å¨åŠ›:")

class DatabaseManager:
    """æ¨¡æ‹Ÿæ•°æ®åº“ç®¡ç†å™¨"""
    def __init__(self):
        # æ¯ä¸ªç±»éƒ½æœ‰è‡ªå·±çš„loggerï¼Œä¾¿äºè¿½è¸ª
        self.logger = logging.getLogger(f"{__name__}.{self.__class__.__name__}")
    
    def connect(self):
        self.logger.info("æ•°æ®åº“è¿æ¥å·²å»ºç«‹")
        
    def query(self, sql):
        self.logger.debug(f"æ‰§è¡ŒSQL: {sql}")
        
    def close(self):
        self.logger.warning("æ•°æ®åº“è¿æ¥å·²å…³é—­")

class APIHandler:
    """æ¨¡æ‹ŸAPIå¤„ç†å™¨"""
    def __init__(self):
        self.logger = logging.getLogger(f"{__name__}.{self.__class__.__name__}")
    
    def handle_request(self, endpoint):
        self.logger.info(f"å¤„ç†è¯·æ±‚: {endpoint}")

# ä½¿ç”¨ç¤ºä¾‹
db = DatabaseManager()
api = APIHandler()

db.connect()
db.query("SELECT * FROM users WHERE active = TRUE")
api.handle_request("/api/users/profile")
db.close()

# å¯¹æ¯”ï¼šä¸ä½¿ç”¨ getLogger çš„é—®é¢˜
print(f"\n{'='*60}")
print("âŒ ä¸ä½¿ç”¨ getLogger(__name__) çš„é—®é¢˜:")
print("-" * 40)

# é—®é¢˜æ¼”ç¤ºï¼šéƒ½å« 'root'ï¼Œæ— æ³•åŒºåˆ†æ¥æº
basic_logger = logging.getLogger()  # ä¸ä¼ å‚æ•°ï¼Œé»˜è®¤æ˜¯root
basic_logger.info("è¿™æ˜¯å“ªä¸ªæ¨¡å—çš„æ—¥å¿—ï¼Ÿæ— æ³•åŒºåˆ†ï¼")

# é—®é¢˜æ¼”ç¤ºï¼šæ‰‹åŠ¨å‘½åå®¹æ˜“é‡å¤
manual_logger1 = logging.getLogger('app')
manual_logger2 = logging.getLogger('app')  # é‡åäº†ï¼
manual_logger1.info("æ¨¡å—Açš„æ—¥å¿—")
manual_logger2.info("æ¨¡å—Bçš„æ—¥å¿—")  # çœ‹èµ·æ¥åƒåŒä¸€ä¸ªæ¨¡å—

print("\nâœ… ä½¿ç”¨ getLogger(__name__) çš„ä¼˜åŠ¿:")
print("1. ğŸ·ï¸  è‡ªåŠ¨å‘½åï¼šæ¨¡å—è·¯å¾„ä½œä¸ºloggerå")
print("2. ğŸŒ³ å±‚æ¬¡ç®¡ç†ï¼šçˆ¶å­loggerç»§æ‰¿å…³ç³»")
print("3. ğŸ” ä¾¿äºè¿½è¸ªï¼šæ—¥å¿—æ˜¾ç¤ºç¡®åˆ‡æ¥æº")
print("4. âš™ï¸  ç»†ç²’åº¦æ§åˆ¶ï¼šä¸åŒæ¨¡å—ä¸åŒçº§åˆ«")
print("5. ğŸ”§ ä¾¿äºè°ƒè¯•ï¼šå¿«é€Ÿå®šä½é—®é¢˜æ¨¡å—")

print(f"\n{'='*60}")
print("ğŸ“ æ€»ç»“ï¼šç”Ÿäº§ç¯å¢ƒæœ€ä½³å®è·µ")
print("="*60)
print("""
1. level é…ç½®ï¼š
   âœ… å¼€å‘ç¯å¢ƒï¼šlogging.DEBUG (çœ‹åˆ°æ‰€æœ‰ä¿¡æ¯)
   âœ… æµ‹è¯•ç¯å¢ƒï¼šlogging.INFO (é‡è¦ä¿¡æ¯)
   âœ… ç”Ÿäº§ç¯å¢ƒï¼šlogging.WARNING (åªçœ‹é—®é¢˜)

2. getLogger(__name__) åŸå› ï¼š
   âœ… è‡ªåŠ¨æ¨¡å—åŒ–å‘½å
   âœ… æ”¯æŒå±‚æ¬¡åŒ–ç®¡ç†
   âœ… ä¾¿äºé—®é¢˜è¿½è¸ª
   âœ… ä»£ç å¯ç»´æŠ¤æ€§å¼º

3. æ ¼å¼åŒ–å»ºè®®ï¼š
   âœ… å¼€å‘ï¼šç®€æ´ + å½©è‰²
   âœ… ç”Ÿäº§ï¼šè¯¦ç»† + JSON (ä¾¿äºåˆ†æ)
   âœ… è°ƒè¯•ï¼šåŒ…å«è¡Œå·å’Œè¿›ç¨‹ID
""") 
```
