```python
#!/usr/bin/env python3
"""
Python Logging æ·±åº¦æŒ‡å—
PyBackendPro ä¸“ä¸šçº§æ—¥å¿—ç®¡ç†æ•™ç¨‹

è§£ç­”æ ¸å¿ƒé—®é¢˜ï¼š
1. basicConfig çš„å¤šæ ·åŒ–é…ç½®é€‰é¡¹
2. getLogger çš„è®¾è®¡ç†å¿µå’Œ __name__ å‚æ•°çš„é‡è¦æ€§
"""

import logging
import sys
from datetime import datetime
from typing import Optional
import os
from pathlib import Path


class LoggingMasterClass:
    """
    æ—¥å¿—ç®¡ç†å¤§å¸ˆè¯¾ - ç”Ÿäº§ç¯å¢ƒçº§åˆ«çš„æ—¥å¿—é…ç½®
    """
    
    def __init__(self):
        self.demo_count = 0
    
    def demo_header(self, title: str):
        """æ¼”ç¤ºåŒºå—æ ‡é¢˜"""
        self.demo_count += 1
        print(f"\n{'='*60}")
        print(f"æ¼”ç¤º {self.demo_count}: {title}")
        print('='*60)
    
    def demonstrate_logging_levels(self):
        """æ¼”ç¤ºæ‰€æœ‰æ—¥å¿—çº§åˆ«"""
        self.demo_header("æ—¥å¿—çº§åˆ«å®Œæ•´è§£æ")
        
        print("""
æ—¥å¿—çº§åˆ«å±‚æ¬¡ç»“æ„ï¼ˆä»ä½åˆ°é«˜ï¼‰ï¼š
NOTSET   = 0   # æœªè®¾ç½®ï¼ˆç»§æ‰¿çˆ¶çº§ï¼‰
DEBUG    = 10  # è°ƒè¯•ä¿¡æ¯ï¼Œå¼€å‘é˜¶æ®µä½¿ç”¨
INFO     = 20  # ä¸€èˆ¬ä¿¡æ¯ï¼Œç¨‹åºæ­£å¸¸è¿è¡Œ
WARNING  = 30  # è­¦å‘Šä¿¡æ¯ï¼Œå¯èƒ½æœ‰é—®é¢˜ä½†ä¸å½±å“è¿è¡Œ
ERROR    = 40  # é”™è¯¯ä¿¡æ¯ï¼ŒåŠŸèƒ½å—å½±å“ä½†ç¨‹åºç»§ç»­
CRITICAL = 50  # ä¸¥é‡é”™è¯¯ï¼Œç¨‹åºå¯èƒ½å´©æºƒ

å…³é”®ç†è§£ï¼šè®¾ç½® level=INFO æ„å‘³ç€åªæ˜¾ç¤º INFOã€WARNINGã€ERRORã€CRITICAL
                      è€Œ DEBUG çº§åˆ«çš„æ¶ˆæ¯ä¼šè¢«å¿½ç•¥
        """)
        
        # æ¼”ç¤ºä¸åŒçº§åˆ«çš„é…ç½®
        levels = [
            ('DEBUG', logging.DEBUG),
            ('INFO', logging.INFO), 
            ('WARNING', logging.WARNING),
            ('ERROR', logging.ERROR),
            ('CRITICAL', logging.CRITICAL)
        ]
        
        for level_name, level_value in levels:
            print(f"\n>>> é…ç½®ä¸º {level_name} çº§åˆ« (å€¼={level_value}):")
            
            # é‡æ–°é…ç½®loggingï¼ˆå¼ºåˆ¶è¦†ç›–ï¼‰
            for handler in logging.root.handlers[:]:
                logging.root.removeHandler(handler)
            
            logging.basicConfig(
                level=level_value,
                format='[%(levelname)8s] %(message)s',
                force=True  # Python 3.8+ æ”¯æŒå¼ºåˆ¶é‡æ–°é…ç½®
            )
            
            logger = logging.getLogger('level_demo')
            
            # æµ‹è¯•æ‰€æœ‰çº§åˆ«çš„è¾“å‡º
            logger.debug("ğŸ› è¿™æ˜¯DEBUGæ¶ˆæ¯ - ç”¨äºè°ƒè¯•ç»†èŠ‚")
            logger.info("â„¹ï¸  è¿™æ˜¯INFOæ¶ˆæ¯ - ç¨‹åºæ­£å¸¸è¿è¡Œ")
            logger.warning("âš ï¸  è¿™æ˜¯WARNINGæ¶ˆæ¯ - éœ€è¦æ³¨æ„")
            logger.error("âŒ è¿™æ˜¯ERRORæ¶ˆæ¯ - å‡ºç°é”™è¯¯")
            logger.critical("ğŸ’¥ è¿™æ˜¯CRITICALæ¶ˆæ¯ - ä¸¥é‡é—®é¢˜")
    
    def demonstrate_basicconfig_options(self):
        """æ¼”ç¤º basicConfig çš„ä¸°å¯Œé…ç½®é€‰é¡¹"""
        self.demo_header("basicConfig é…ç½®é€‰é¡¹å¤§å…¨")
        
        print("""
basicConfig æ”¯æŒçš„ä¸»è¦å‚æ•°ï¼š
- level: æ—¥å¿—çº§åˆ«
- format: è¾“å‡ºæ ¼å¼
- datefmt: æ—¶é—´æ ¼å¼
- filename: è¾“å‡ºåˆ°æ–‡ä»¶
- filemode: æ–‡ä»¶æ¨¡å¼ ('a'è¿½åŠ , 'w'è¦†ç›–)
- stream: è¾“å‡ºæµ (å¦‚ sys.stdout, sys.stderr)
- handlers: è‡ªå®šä¹‰å¤„ç†å™¨åˆ—è¡¨
- force: å¼ºåˆ¶é‡æ–°é…ç½® (Python 3.8+)
        """)
        
        # 1. åŸºç¡€æ ¼å¼é…ç½®
        print("\n1ï¸âƒ£ åŸºç¡€æ ¼å¼é…ç½®:")
        logging.basicConfig(
            level=logging.INFO,
            format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
            datefmt='%Y-%m-%d %H:%M:%S',
            force=True
        )
        logger = logging.getLogger('basic_demo')
        logger.info("è¿™æ˜¯åŸºç¡€æ ¼å¼çš„æ—¥å¿—æ¶ˆæ¯")
        
        # 2. è¯¦ç»†æ ¼å¼é…ç½®
        print("\n2ï¸âƒ£ ç”Ÿäº§ç¯å¢ƒè¯¦ç»†æ ¼å¼:")
        logging.basicConfig(
            level=logging.DEBUG,
            format='[%(asctime)s] [PID:%(process)d] [%(name)s:%(lineno)d] [%(levelname)s] %(message)s',
            datefmt='%Y-%m-%d %H:%M:%S',
            force=True
        )
        logger = logging.getLogger('production_demo')
        logger.info("ç”Ÿäº§ç¯å¢ƒæ ¼å¼ - åŒ…å«è¿›ç¨‹IDå’Œè¡Œå·")
        logger.error("é”™è¯¯æ¶ˆæ¯ç¤ºä¾‹")
        
        # 3. ç®€æ´æ ¼å¼
        print("\n3ï¸âƒ£ ç®€æ´æ ¼å¼:")
        logging.basicConfig(
            level=logging.INFO,
            format='%(levelname)s: %(message)s',
            force=True
        )
        logger = logging.getLogger('simple_demo')
        logger.warning("ç®€æ´æ ¼å¼çš„è­¦å‘Šæ¶ˆæ¯")
        
        # 4. æ–‡ä»¶è¾“å‡ºé…ç½®
        print("\n4ï¸âƒ£ æ–‡ä»¶è¾“å‡ºé…ç½®:")
        log_file = Path("demo_app.log")
        logging.basicConfig(
            level=logging.DEBUG,
            format='%(asctime)s [%(levelname)s] %(name)s: %(message)s',
            filename=log_file,
            filemode='w',  # è¦†ç›–æ¨¡å¼
            force=True
        )
        file_logger = logging.getLogger('file_demo')
        file_logger.info("è¿™æ¡æ¶ˆæ¯å†™å…¥åˆ°æ–‡ä»¶")
        file_logger.error("æ–‡ä»¶ä¸­çš„é”™è¯¯æ¶ˆæ¯")
        
        if log_file.exists():
            print(f"âœ… æ—¥å¿—å·²å†™å…¥æ–‡ä»¶: {log_file}")
            print("æ–‡ä»¶å†…å®¹:")
            with open(log_file, 'r', encoding='utf-8') as f:
                for line in f:
                    print(f"    {line.strip()}")
            log_file.unlink()  # æ¸…ç†æ¼”ç¤ºæ–‡ä»¶
    
    def demonstrate_colored_logging(self):
        """æ¼”ç¤ºå½©è‰²æ—¥å¿—è¾“å‡º"""
        self.demo_header("å½©è‰²æ—¥å¿—è¾“å‡º (éœ€è¦ colorlog åº“)")
        
        try:
            import colorlog
            
            # é…ç½®å½©è‰²æ—¥å¿—
            color_formatter = colorlog.ColoredFormatter(
                '%(log_color)s[%(asctime)s] [%(levelname)8s] %(name)s: %(message)s',
                datefmt='%H:%M:%S',
                log_colors={
                    'DEBUG': 'cyan',
                    'INFO': 'green', 
                    'WARNING': 'yellow',
                    'ERROR': 'red',
                    'CRITICAL': 'red,bg_white',
                }
            )
            
            # åˆ›å»ºå½©è‰²å¤„ç†å™¨
            handler = colorlog.StreamHandler()
            handler.setFormatter(color_formatter)
            
            # é…ç½®logger
            color_logger = colorlog.getLogger('color_demo')
            color_logger.setLevel(logging.DEBUG)
            color_logger.handlers.clear()
            color_logger.addHandler(handler)
            
            print("ğŸŒˆ å½©è‰²æ—¥å¿—æ¼”ç¤º:")
            color_logger.debug("è°ƒè¯•ä¿¡æ¯ (é’è‰²)")
            color_logger.info("æ­£å¸¸ä¿¡æ¯ (ç»¿è‰²)")
            color_logger.warning("è­¦å‘Šä¿¡æ¯ (é»„è‰²)")
            color_logger.error("é”™è¯¯ä¿¡æ¯ (çº¢è‰²)")
            color_logger.critical("ä¸¥é‡é”™è¯¯ (çº¢è‰²èƒŒæ™¯ç™½å­—)")
            
        except ImportError:
            print("âŒ éœ€è¦å®‰è£… colorlog: pip install colorlog")
            print("ğŸ’¡ æ ‡å‡†åº“æ›¿ä»£æ–¹æ¡ˆ:")
            
            # ä½¿ç”¨ANSIè½¬ä¹‰åºåˆ—çš„ç®€å•å½©è‰²è¾“å‡º
            class ColoredFormatter(logging.Formatter):
                """ç®€å•çš„å½©è‰²æ ¼å¼å™¨"""
                
                COLORS = {
                    'DEBUG': '\033[36m',    # é’è‰²
                    'INFO': '\033[32m',     # ç»¿è‰²
                    'WARNING': '\033[33m',  # é»„è‰²
                    'ERROR': '\033[31m',    # çº¢è‰²
                    'CRITICAL': '\033[41m', # çº¢è‰²èƒŒæ™¯
                    'RESET': '\033[0m'      # é‡ç½®
                }
                
                def format(self, record):
                    color = self.COLORS.get(record.levelname, '')
                    reset = self.COLORS['RESET']
                    record.levelname = f"{color}{record.levelname}{reset}"
                    return super().format(record)
            
            # é…ç½®ç®€å•å½©è‰²æ—¥å¿—
            handler = logging.StreamHandler()
            handler.setFormatter(ColoredFormatter(
                '[%(asctime)s] [%(levelname)s] %(name)s: %(message)s',
                datefmt='%H:%M:%S'
            ))
            
            simple_color_logger = logging.getLogger('simple_color')
            simple_color_logger.setLevel(logging.DEBUG)
            simple_color_logger.handlers.clear()
            simple_color_logger.addHandler(handler)
            
            simple_color_logger.debug("è°ƒè¯•ä¿¡æ¯")
            simple_color_logger.info("æ­£å¸¸ä¿¡æ¯") 
            simple_color_logger.warning("è­¦å‘Šä¿¡æ¯")
            simple_color_logger.error("é”™è¯¯ä¿¡æ¯")
            simple_color_logger.critical("ä¸¥é‡é”™è¯¯")
    
    def explain_getlogger_and_name(self):
        """æ·±åº¦è§£æ getLogger å’Œ __name__ çš„è®¾è®¡ç†å¿µ"""
        self.demo_header("getLogger å’Œ __name__ çš„æ ¸å¿ƒç†å¿µ")
        
        print("""
ğŸ¯ ä¸ºä»€ä¹ˆä½¿ç”¨ logging.getLogger(__name__) ï¼Ÿ

1. **æ¨¡å—åŒ–æ—¥å¿—ç®¡ç†**ï¼š
   - æ¯ä¸ªæ¨¡å—æœ‰ç‹¬ç«‹çš„loggerå®ä¾‹
   - ä¾¿äºè¿½è¸ªæ—¥å¿—æ¥æº
   - æ”¯æŒç»†ç²’åº¦çš„æ—¥å¿—æ§åˆ¶

2. **__name__ çš„å¦™ç”¨**ï¼š
   - __name__ æ˜¯æ¨¡å—çš„å®Œæ•´è·¯å¾„å
   - å½¢æˆå±‚æ¬¡åŒ–çš„loggeræ ‘ç»“æ„
   - æ”¯æŒç»§æ‰¿å¼é…ç½®

3. **Logger å±‚æ¬¡ç»“æ„**ï¼š
   - ç±»ä¼¼åŒ…çš„å±‚æ¬¡ç»“æ„
   - å­loggerç»§æ‰¿çˆ¶loggerçš„é…ç½®
   - æ ¹loggeræ˜¯æ‰€æœ‰loggerçš„ç¥–å…ˆ
        """)
        
        # æ¼”ç¤ºæ¨¡å—åçš„å®é™…å€¼
        print(f"\nğŸ“ å½“å‰æ¨¡å—çš„ __name__ = '{__name__}'")
        print(f"ğŸ“ å¦‚æœæ˜¯ä¸»ç¨‹åºï¼Œ__name__ = '__main__'")
        print(f"ğŸ“ å¦‚æœæ˜¯å¯¼å…¥çš„æ¨¡å—ï¼Œ__name__ = 'æ¨¡å—çš„å®Œæ•´è·¯å¾„'")
        
        # æ¼”ç¤ºloggerå±‚æ¬¡ç»“æ„
        print("\nğŸŒ³ Logger å±‚æ¬¡ç»“æ„æ¼”ç¤º:")
        
        # åˆ›å»ºå±‚æ¬¡åŒ–çš„logger
        root_logger = logging.getLogger()
        app_logger = logging.getLogger('myapp')
        db_logger = logging.getLogger('myapp.database')
        api_logger = logging.getLogger('myapp.api')
        auth_logger = logging.getLogger('myapp.api.auth')
        
        print(f"æ ¹logger: {root_logger.name} (çº§åˆ«: {root_logger.level})")
        print(f"åº”ç”¨logger: {app_logger.name}")
        print(f"æ•°æ®åº“logger: {db_logger.name}")
        print(f"API logger: {api_logger.name}")
        print(f"è®¤è¯logger: {auth_logger.name}")
        
        # é…ç½®çˆ¶çº§logger
        logging.basicConfig(
            level=logging.INFO,
            format='[%(name)20s] [%(levelname)8s] %(message)s',
            force=True
        )
        
        # è®¾ç½®ç‰¹å®šæ¨¡å—çš„æ—¥å¿—çº§åˆ«
        app_logger.setLevel(logging.DEBUG)
        db_logger.setLevel(logging.WARNING)  # æ•°æ®åº“æ—¥å¿—åªæ˜¾ç¤ºè­¦å‘Šä»¥ä¸Š
        
        print("\nğŸ“Š ç»§æ‰¿å…³ç³»æµ‹è¯•:")
        app_logger.debug("åº”ç”¨è°ƒè¯•ä¿¡æ¯")
        app_logger.info("åº”ç”¨æ­£å¸¸ä¿¡æ¯")
        
        db_logger.debug("æ•°æ®åº“è°ƒè¯• (ä¸ä¼šæ˜¾ç¤º)")
        db_logger.info("æ•°æ®åº“ä¿¡æ¯ (ä¸ä¼šæ˜¾ç¤º)")
        db_logger.warning("æ•°æ®åº“è­¦å‘Š (ä¼šæ˜¾ç¤º)")
        
        api_logger.info("APIä¿¡æ¯")
        auth_logger.info("è®¤è¯ä¿¡æ¯")
    
    def demonstrate_advanced_logger_usage(self):
        """æ¼”ç¤ºé«˜çº§loggerä½¿ç”¨æŠ€å·§"""
        self.demo_header("ç”Ÿäº§ç¯å¢ƒ Logger æœ€ä½³å®è·µ")
        
        # 1. å¤šå¤„ç†å™¨é…ç½®
        print("1ï¸âƒ£ å¤šå¤„ç†å™¨é…ç½® (åŒæ—¶è¾“å‡ºåˆ°æ§åˆ¶å°å’Œæ–‡ä»¶):")
        
        # åˆ›å»ºä¸“ä¸šçš„logger
        professional_logger = logging.getLogger('professional_app')
        professional_logger.setLevel(logging.DEBUG)
        
        # æ¸…é™¤ç°æœ‰å¤„ç†å™¨
        professional_logger.handlers.clear()
        
        # æ§åˆ¶å°å¤„ç†å™¨ - åªæ˜¾ç¤º INFO ä»¥ä¸Š
        console_handler = logging.StreamHandler(sys.stdout)
        console_handler.setLevel(logging.INFO)
        console_formatter = logging.Formatter(
            'ğŸ–¥ï¸  [%(asctime)s] [%(levelname)8s] %(name)s: %(message)s',
            datefmt='%H:%M:%S'
        )
        console_handler.setFormatter(console_formatter)
        
        # æ–‡ä»¶å¤„ç†å™¨ - è®°å½•æ‰€æœ‰çº§åˆ«
        file_handler = logging.FileHandler('professional_app.log', mode='w')
        file_handler.setLevel(logging.DEBUG)
        file_formatter = logging.Formatter(
            '[%(asctime)s] [PID:%(process)d] [%(name)s:%(lineno)d] [%(levelname)s] %(message)s',
            datefmt='%Y-%m-%d %H:%M:%S'
        )
        file_handler.setFormatter(file_formatter)
        
        # æ·»åŠ å¤„ç†å™¨
        professional_logger.addHandler(console_handler)
        professional_logger.addHandler(file_handler)
        
        # é˜²æ­¢é‡å¤è¾“å‡º
        professional_logger.propagate = False
        
        # æµ‹è¯•è¾“å‡º
        professional_logger.debug("è°ƒè¯•ä¿¡æ¯ (ä»…æ–‡ä»¶)")
        professional_logger.info("æ­£å¸¸ä¿¡æ¯ (æ§åˆ¶å°+æ–‡ä»¶)")
        professional_logger.warning("è­¦å‘Šä¿¡æ¯ (æ§åˆ¶å°+æ–‡ä»¶)")
        professional_logger.error("é”™è¯¯ä¿¡æ¯ (æ§åˆ¶å°+æ–‡ä»¶)")
        
        print("\nğŸ“ æŸ¥çœ‹æ–‡ä»¶å†…å®¹:")
        if Path('professional_app.log').exists():
            with open('professional_app.log', 'r') as f:
                for line in f:
                    print(f"    {line.strip()}")
            Path('professional_app.log').unlink()
        
        # 2. æ¡ä»¶æ—¥å¿—è®°å½•
        print("\n2ï¸âƒ£ æ¡ä»¶æ—¥å¿—å’Œæ€§èƒ½ä¼˜åŒ–:")
        
        perf_logger = logging.getLogger('performance')
        perf_logger.setLevel(logging.INFO)
        
        # âŒ ä½æ•ˆæ–¹å¼ - æ€»æ˜¯è®¡ç®—å­—ç¬¦ä¸²
        expensive_data = {"key": "value" * 1000}
        # perf_logger.debug(f"æ˜‚è´µçš„è°ƒè¯•ä¿¡æ¯: {expensive_data}")  # å³ä½¿ä¸è¾“å‡ºä¹Ÿä¼šè®¡ç®—
        
        # âœ… é«˜æ•ˆæ–¹å¼ - æ¡ä»¶è®¡ç®—
        if perf_logger.isEnabledFor(logging.DEBUG):
            perf_logger.debug(f"æ˜‚è´µçš„è°ƒè¯•ä¿¡æ¯: {expensive_data}")
        
        # âœ… æ›´ä¼˜é›…çš„æ–¹å¼ - å»¶è¿Ÿæ ¼å¼åŒ–
        perf_logger.debug("å»¶è¿Ÿæ ¼å¼åŒ–: %s", lambda: expensive_data)
        
        print("âœ… æ€§èƒ½ä¼˜åŒ–æŠ€å·§å·²æ¼”ç¤º")
    
    def demonstrate_structured_logging(self):
        """æ¼”ç¤ºç»“æ„åŒ–æ—¥å¿—"""
        self.demo_header("ç»“æ„åŒ–æ—¥å¿— - ç°ä»£æ—¥å¿—ç®¡ç†")
        
        try:
            import structlog
            
            # é…ç½®structlog
            structlog.configure(
                processors=[
                    structlog.stdlib.filter_by_level,
                    structlog.stdlib.add_logger_name,
                    structlog.stdlib.add_log_level,
                    structlog.processors.TimeStamper(fmt="iso"),
                    structlog.dev.ConsoleRenderer()
                ],
                context_class=dict,
                logger_factory=structlog.stdlib.LoggerFactory(),
                wrapper_class=structlog.stdlib.BoundLogger,
                cache_logger_on_first_use=True,
            )
            
            logger = structlog.get_logger("structured_demo")
            
            print("ğŸ—ï¸ ç»“æ„åŒ–æ—¥å¿—æ¼”ç¤º:")
            logger.info("ç”¨æˆ·ç™»å½•", user_id=12345, ip="192.168.1.1", method="POST")
            logger.warning("APIè°ƒç”¨ç¼“æ…¢", endpoint="/api/users", duration=2.5, threshold=2.0)
            logger.error("æ•°æ®åº“è¿æ¥å¤±è´¥", 
                        host="db.example.com", 
                        port=5432, 
                        error_code="CONNECTION_TIMEOUT",
                        retry_count=3)
            
        except ImportError:
            print("âŒ éœ€è¦å®‰è£… structlog: pip install structlog")
            print("ğŸ’¡ æ ‡å‡†åº“æ›¿ä»£æ–¹æ¡ˆ:")
            
            # ä½¿ç”¨æ ‡å‡†åº“å®ç°ç»“æ„åŒ–æ—¥å¿—
            class StructuredLogger:
                def __init__(self, name: str):
                    self.logger = logging.getLogger(name)
                
                def info(self, message: str, **kwargs):
                    extra_info = " | ".join(f"{k}={v}" for k, v in kwargs.items())
                    self.logger.info(f"{message} | {extra_info}")
                
                def error(self, message: str, **kwargs):
                    extra_info = " | ".join(f"{k}={v}" for k, v in kwargs.items())
                    self.logger.error(f"{message} | {extra_info}")
            
            struct_logger = StructuredLogger("manual_structured")
            struct_logger.info("ç”¨æˆ·æ“ä½œ", user_id=67890, action="delete_file", file_size="2.5MB")
            struct_logger.error("æ”¯ä»˜å¤±è´¥", order_id="ORD-123", amount=99.99, payment_method="credit_card")


def main():
    """ä¸»å‡½æ•° - è¿è¡Œæ‰€æœ‰æ¼”ç¤º"""
    print("ğŸ“ PyBackendPro Logging æ·±åº¦æŒ‡å—")
    print("="*60)
    
    master = LoggingMasterClass()
    
    # 1. åŸºç¡€çŸ¥è¯†
    master.demonstrate_logging_levels()
    master.demonstrate_basicconfig_options()
    
    # 2. è¿›é˜¶åŠŸèƒ½
    master.demonstrate_colored_logging()
    master.explain_getlogger_and_name()
    
    # 3. ç”Ÿäº§å®è·µ
    master.demonstrate_advanced_logger_usage()
    master.demonstrate_structured_logging()
    
    print(f"\n{'='*60}")
    print("ğŸ¯ æ ¸å¿ƒè¦ç‚¹æ€»ç»“:")
    print("1. level å‚æ•°æ§åˆ¶æ˜¾ç¤ºå“ªäº›çº§åˆ«çš„æ—¥å¿—")
    print("2. basicConfig æ”¯æŒä¸°å¯Œçš„æ ¼å¼åŒ–é€‰é¡¹")
    print("3. getLogger(__name__) åˆ›å»ºæ¨¡å—åŒ–çš„æ—¥å¿—ç®¡ç†")
    print("4. __name__ æä¾›äº†å¤©ç„¶çš„å±‚æ¬¡ç»“æ„")
    print("5. ç”Ÿäº§ç¯å¢ƒå»ºè®®ä½¿ç”¨å¤šå¤„ç†å™¨é…ç½®")
    print("6. ç»“æ„åŒ–æ—¥å¿—æ˜¯ç°ä»£åº”ç”¨çš„æœ€ä½³å®è·µ")
    print("="*60)


if __name__ == "__main__":
    main() 
```
