## ğŸ” åŸºæœ¬æ“ä½œæ–¹æ³•

### æ·»åŠ å…ƒç´ 
| æ–¹æ³• | åŠŸèƒ½ | ç”¨æ³• |
|------|------|------|
| `add(elem)` | æ·»åŠ å•ä¸ªå…ƒç´  | `s.add(5)` |
| `update(iterable)` | æ·»åŠ å¤šä¸ªå…ƒç´  | `s.update([4, 5, 6])` |

### åˆ é™¤å…ƒç´   
| æ–¹æ³• | åŠŸèƒ½ | åŒºåˆ« |
|------|------|------|
| `remove(elem)` | åˆ é™¤æŒ‡å®šå…ƒç´  | å…ƒç´ ä¸å­˜åœ¨ä¼šæŠ¥é”™ |
| `discard(elem)` | åˆ é™¤æŒ‡å®šå…ƒç´  | å…ƒç´ ä¸å­˜åœ¨ä¸æŠ¥é”™ |
| `pop()` | åˆ é™¤å¹¶è¿”å›ä»»æ„å…ƒç´  | é›†åˆä¸ºç©ºä¼šæŠ¥é”™ |
| `clear()` | æ¸…ç©ºé›†åˆ | é›†åˆå˜ä¸ºç©ºé›†åˆ |

### æŸ¥è¯¢æ“ä½œ
| æ–¹æ³• | åŠŸèƒ½ | è¿”å›å€¼ |
|------|------|--------|
| `len(s)` | è·å–å…ƒç´ ä¸ªæ•° | æ•´æ•° |
| `elem in s` | æ£€æŸ¥å…ƒç´ æ˜¯å¦å­˜åœ¨ | True/False |
| `elem not in s` | æ£€æŸ¥å…ƒç´ æ˜¯å¦ä¸å­˜åœ¨ | True/False |


````python
# åŸºæœ¬æ“ä½œæ¼”ç¤º
s = {1, 2, 3}
print(f"åˆå§‹é›†åˆ: {s}")

# æ·»åŠ å…ƒç´ 
s.add(4)
print(f"add(4)å: {s}")

s.update([5, 6, 7])
print(f"update([5,6,7])å: {s}")

# åˆ é™¤å…ƒç´ 
s.remove(7)  # åˆ é™¤7
print(f"remove(7)å: {s}")

s.discard(10)  # åˆ é™¤ä¸å­˜åœ¨çš„å…ƒç´ ï¼Œä¸ä¼šæŠ¥é”™
print(f"discard(10)å: {s}")

# å¼¹å‡ºå…ƒç´ 
popped = s.pop()
print(f"pop()è¿”å›: {popped}, å‰©ä½™: {s}")

# æŸ¥è¯¢æ“ä½œ
print(f"é•¿åº¦: {len(s)}")
print(f"2 in s: {2 in s}")
print(f"10 not in s: {10 not in s}")
````

## ğŸ§® é›†åˆè¿ç®—æ–¹æ³•

### æ•°å­¦é›†åˆæ“ä½œ
| è¿ç®— | æ–¹æ³•å½¢å¼ | è¿ç®—ç¬¦å½¢å¼ | åŠŸèƒ½ |
|------|----------|------------|------|
| **å¹¶é›†** | `s1.union(s2)` | `s1 \| s2` | ä¸¤é›†åˆæ‰€æœ‰å…ƒç´  |
| **äº¤é›†** | `s1.intersection(s2)` | `s1 & s2` | ä¸¤é›†åˆå…±åŒå…ƒç´  |
| **å·®é›†** | `s1.difference(s2)` | `s1 - s2` | s1ä¸­æœ‰ä½†s2æ²¡æœ‰çš„å…ƒç´  |
| **å¯¹ç§°å·®é›†** | `s1.symmetric_difference(s2)` | `s1 ^ s2` | ä¸¤é›†åˆä¸å…±åŒçš„å…ƒç´  |

### é›†åˆå…³ç³»åˆ¤æ–­
| æ–¹æ³• | è¿ç®—ç¬¦ | åŠŸèƒ½ |
|------|--------|------|
| `s1.issubset(s2)` | `s1 <= s2` | s1æ˜¯å¦ä¸ºs2çš„å­é›† |
| `s1.issuperset(s2)` | `s1 >= s2` | s1æ˜¯å¦ä¸ºs2çš„è¶…é›† |
| `s1.isdisjoint(s2)` | æ—  | ä¸¤é›†åˆæ˜¯å¦æ— äº¤é›† |

### åŸåœ°æ›´æ–°æ–¹æ³•
| æ–¹æ³• | è¿ç®—ç¬¦ | åŠŸèƒ½ |
|------|--------|------|
| `s1.update(s2)` | `s1 \|= s2` | å°†s2å¹¶å…¥s1 |
| `s1.intersection_update(s2)` | `s1 &= s2` | s1ä¿ç•™ä¸s2çš„äº¤é›† |
| `s1.difference_update(s2)` | `s1 -= s2` | ä»s1ä¸­åˆ é™¤s2çš„å…ƒç´  |
| `s1.symmetric_difference_update(s2)` | `s1 ^= s2` | s1æ›´æ–°ä¸ºå¯¹ç§°å·®é›† |


````python
# é›†åˆè¿ç®—æ¼”ç¤º
s1 = {1, 2, 3, 4}
s2 = {3, 4, 5, 6}

print(f"s1: {s1}")
print(f"s2: {s2}")
print()

# é›†åˆè¿ç®—
print("=== é›†åˆè¿ç®— ===")
print(f"å¹¶é›† s1 | s2: {s1 | s2}")
print(f"äº¤é›† s1 & s2: {s1 & s2}")
print(f"å·®é›† s1 - s2: {s1 - s2}")
print(f"å¯¹ç§°å·®é›† s1 ^ s2: {s1 ^ s2}")
print()

# é›†åˆå…³ç³»
print("=== é›†åˆå…³ç³» ===")
s3 = {1, 2}
print(f"s3: {s3}")
print(f"s3æ˜¯s1çš„å­é›†: {s3 <= s1}")
print(f"s1æ˜¯s3çš„è¶…é›†: {s1 >= s3}")
print(f"s1å’Œs2æ— äº¤é›†: {s1.isdisjoint(s2)}")
print()

# åŸåœ°æ›´æ–°æ¼”ç¤º
print("=== åŸåœ°æ›´æ–° ===")
s_copy = s1.copy()
print(f"åŸå§‹: {s_copy}")
s_copy |= {7, 8}
print(f"å¹¶å…¥{7,8}å: {s_copy}")
````

````python
# å¸¸è§ç”¨é€”æ¼”ç¤º
import time

# 1. æ•°æ®å»é‡
print("=== æ•°æ®å»é‡ ===")
numbers = [1, 2, 2, 3, 3, 4, 4, 5]
unique_numbers = list(set(numbers))
print(f"åŸåˆ—è¡¨: {numbers}")
print(f"å»é‡å: {unique_numbers}")
print()

# 2. é«˜æ•ˆæˆå‘˜æµ‹è¯•
print("=== æ€§èƒ½å¯¹æ¯” ===")
large_list = list(range(100000))
large_set = set(large_list)

# æµ‹è¯•æŸ¥æ‰¾æ€§èƒ½
target = 99999

# åˆ—è¡¨æŸ¥æ‰¾
start = time.time()
result1 = target in large_list
time1 = time.time() - start

# é›†åˆæŸ¥æ‰¾
start = time.time()
result2 = target in large_set
time2 = time.time() - start

print(f"åˆ—è¡¨æŸ¥æ‰¾è€—æ—¶: {time1:.6f}ç§’")
print(f"é›†åˆæŸ¥æ‰¾è€—æ—¶: {time2:.6f}ç§’")
print(f"é›†åˆæ¯”åˆ—è¡¨å¿« {time1/time2:.1f} å€")
print()

# 3. é›†åˆè¿ç®—åº”ç”¨
print("=== å®é™…åº”ç”¨ ===")
favorites_alice = {'python', 'java', 'go', 'rust'}
favorites_bob = {'python', 'javascript', 'go', 'c++'}

print(f"Aliceå–œæ¬¢: {favorites_alice}")
print(f"Bobå–œæ¬¢: {favorites_bob}")
print(f"å…±åŒå–œæ¬¢: {favorites_alice & favorites_bob}")
print(f"Aliceç‹¬æœ‰: {favorites_alice - favorites_bob}")
print(f"Bobç‹¬æœ‰: {favorites_bob - favorites_alice}")
print(f"æ‰€æœ‰è¯­è¨€: {favorites_alice | favorites_bob}")
````
